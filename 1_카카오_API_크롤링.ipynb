```python
KAKAO_API_KEY = "내 API 키"
KAKAO_WEB_SEARCH_URL = "https://dapi.kakao.com/v2/search/web?query={}" # 주소 : 카카오-문서-REST API-검색/광고/결제-Daum 검색-지원하는 기능-웹문서 검색하기-URL
```

카카오 api에 요청하기 위해서는 인증 정보가 필요!

```python
header = {
  "Authorization": "KakaoAK {}".format(KAKAO_API_KEY)
}

header
```

```python
# https://developers.kakao.com/docs/latest/ko/daum-search/dev-guide#search-doc GET방식
import requests # 파이썬에서 요청
response = requests.get(KAKAO_WEB_SEARCH_URL.format("스우파2"), headers=header) # 요청 URL 및 헤더 정보 넣어주기 / 스우파2가 KAKAO_WEB_SEARCH_URL = "https://dapi.kakao.com/v2/search/web?query={스우파2}" 이렇게 들어감
response
```

```python
# 응답 코드 가져오기
response.status_code
```

응답 코드(Response Code)
* `1xx` : 처리 중 ( 볼 일이 없다 )
* **`2xx`** : Success
* `3xx` : Redirection ( 추가 동작이 필요하다. - 볼 일이 많이 없다 )
* **`4xx`** : Client Error ( 클라이언트에서 오류 발생 )
    - `404` : Page Not Found (잘못된 url로 요청함)
* **`5xx`** : Server Error ( 서버에서 오류발생 )

```python
KAKAO_API_KEY = "내 API 키"
KAKAO_WEB_SEARCH_URL = "https://dapi.kakao.com/v2/search/web" # 주소 : 카카오-문서-REST API-검색/광고/결제-Daum 검색-지원하는 기능-웹문서 검색하기-URL
```

```python
header = {
  "Authorization": "KakaoAK {}".format(KAKAO_API_KEY)
}

header
```

```python
page = 50  # 페이지 번호
size = 50  # 페이지 크기

# 검색어 설정
query = "스우파2"

# 요청 URL에 파라미터 추가
params = {
    "query": query,
    "page": page,  # 페이지 번호
    "size": size  # 페이지 크기
}
```

```python
import requests # 파이썬에서 요청
response = requests.get(KAKAO_WEB_SEARCH_URL, headers=header, params=params)
response
```

```python
datas = response.json()  # documents로 구성된 JSON 객체 / 다음사이트에서 검색어 웹문서
datas
```

```python
documents = datas['documents']
pd.DataFrame(documents)
```

```python
query = '장원영'  # 검색어 설정
total_results = []  # 모든 결과를 저장할 리스트

for page in range(1, 51):  # 예시로 5페이지까지 크롤링
    size = 50  # 페이지 크기 설정

    # API 요청을 보내고 결과를 파싱
    params = {'query': query, 'page': page, 'size': size}
    response = requests.get(KAKAO_WEB_SEARCH_URL, headers=header, params=params)
    data = response.json()

    # 검색 결과가 있는 경우 결과를 리스트에 추가
    if 'documents' in data:
        results = data['documents']
        total_results.extend(results)

# 크롤링한 데이터 사용 예시
# for result in total_results:
#     print(f'Title: {result["title"]}')
#     print(f'URL: {result["url"]}')
#     print('---')
```
